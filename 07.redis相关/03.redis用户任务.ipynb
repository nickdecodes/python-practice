```python
#!/usr/bin/env python
# coding=utf-8

import redis
import random
import json

pool = redis.ConnectionPool(host='localhost', port=6379, decode_responses=True)
r = redis.Redis(host='localhost', port=6379, decode_responses=True)

def random_name(sex):
    surname = ["赵", "钱", "孙", "李", "周", "吴", "郑", "王", "冯", "陈", "褚", "卫", "蒋", "沈", "韩", "杨", "朱", "秦", "尤", "许", "何", "吕", "施", "张", "孔", "曹", "严", "华", "金", "魏", "陶", "姜"]
    girl = ["秀", "娟", "英", "华", "慧", "巧", "美", "娜", "静", "淑", "惠", "珠", "翠", "雅", "芝", "玉", "萍", "红", "娥", "玲", "芬", "芳", "燕", "彩", "春", "菊", "兰", "凤", "洁", "梅", "琳", "素", "云"]
    boy = ["伟", "刚", "勇", "毅", "俊", "峰", "强", "军", "平", "保", "东", "文", "辉", "力", "明", "永", "健", "世", "广", "志", "义", "兴", "良", "海", "山", "仁", "波", "宁", "贵", "福", "生", "龙"]
    if sex == "女":
        name = random.choice(surname) + random.choice(girl)
    elif sex == "男":
        name = random.choice(surname) + random.choice(boy)
    return name

def random_sex():
    #0 男
    #1 女
    sex_ch = ["男", "女"]
    sex = random.choice(sex_ch)
    return sex

def random_phone():
    phone_area = ["182", "187", "156"]
    phone_num = random.randint(10000000, 99999999)
    phone = random.choice(phone_area) + str(phone_num)
    return phone

def random_age():
    age = random.choice(range(0, 100))
    return age

def save_to_redis():
    for i in range(10001):
        sex = random_sex()
        name = random_name(sex)
        age = random_age()
        phone = random_phone()
        user = "user" + str(random.randint(0, 100000000))
        data = { "name": name, "sex": sex, "age": age, "phone": phone }
        r.hmset(user, data)
        r.sadd("userID", user)
        r.zadd("zset", {user : age})
        
        #print "name : %s, sex : %s, age : %d, phone : %s" % (name, sex, age, phone)

def print_redis():
    for user in r.smembers("userID"):
        data = r.hgetall(user)
        print "name : %s, sex : %s, age : %s, phone : %s" % (data['name'], data['sex'], data['age'], data['phone'])

def search_age(begin, end):
    for user in r.zrangebyscore("zset", begin, end):
        data = r.hgetall(user)
        print "name : %s, sex : %s, age : %s, phone : %s" % (data['name'], data['sex'], data['age'], data['phone'])

def search_info(info, content):
    for user in r.smembers("userID"):
        user = r.hgetall(user)
        if (user[info] == content.decode("utf-8")):
            print "name : %s, sex : %s, age : %s, phone : %s" % (user['name'], user['sex'], user['age'], user['phone'])


"""
if __name__ == "__main__":
    r.hset("1", "name", "zhangsan")
    r.hset("1", "name", "lisi")
    save_to_redis()
    #print_redis()
    #search_age(0,1)
"""
save_to_redis()
print_redis()
```
