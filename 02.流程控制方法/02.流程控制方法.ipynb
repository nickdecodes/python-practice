## 1、分支控制

在人类最基本的控制逻辑中，根据现有条件决定下一步的事情是基本的逻辑之一，我们管这种逻辑叫做 `分支控制`。在 C/C++ 中，我们学过 $if、 else、 $，在 **Python** 也是一样的，下面就让我们来看看 **Python** 中的 `分支控制` 语句。  

从这一课开始，我们将正式接触 **Python** 中代码块的概念，不同于 C/C++ 中用 $\{\}$ 标记代码块，**Python** 中是用缩进来控制代码块的，连续的相同缩进语句，将组成一个代码块，不理解，不用怕，一起跟船长来看看吧。

```python
# example 2.1.1

# 定义一个比较函数
# 函数内容后面会具体讲到
def cmp(a, b):
    print ("cmp(%d, %d)" %(a, b))
    if a == b:
        print ("\ta == b is true")
        print ("\ta == b == %d" % (a))
    elif abs(a - b) == 1:
        print ("\ta == b is false")
        print ("\ta == %d, b == %d" % (a, b))
    else:
        print ("\ta == b is false")
        print ("\t|a - b| >= 2")

cmp(1, 1) # 测试 if 分支
cmp(1, 2) # 测试 elif 分支
cmp(1, 3) # 测试 else 分支
```

## 2、循环控制

上面看完 `分支控制` 之后，下面让我们来看看 `循环控制`，当我们需要将同样的流程执行多次的时候，这个时候就需要 `循环` 出马了。  
下面在介绍循环的过程中，会引入两个重要的 **Python** 功能函数： $range(), xrange()$
1. `range(a, b)` 生成一个包含 \[a, b) 区间中所有元素的列表（暂时可以理解为数组）
2. `xrange(a, b)` 生成一个用于生成 \[a, b) 区间中所有元素的迭代器（暂时理解为【高效的 range 函数】）

```python
# example 2.2.1
# 利用 range 从 1 到 4(5 - 1)  循环
for i in range(1, 5):
    print (i)
    continue
else :
    print (" this is end")

# 利用 xrange 从 1 到 4(5 - 1) 循环
for i in range(1, 5):
    print (i)
    if i == 3:
        break
else :
    print (" this is end")
```

以上演示了 `for` 循环的相关功能，下面来让我们看看 `while` 循环的功能。当然，`while` 里面使用 ***break*** 或者 ***continue*** 和 `for` 里面的效果是一样的，这个就不带大家具体看了，小伙伴们线下自己试验吧。

```python
# example 2.2.2
i = 10
while i:
    print ("%d" % (i % 2))
    i /= 2
else :
    print ("| finally : i = %d" % i)
```

### 3.2 代码

```python
def is_prime(x):
    if type(x) != type(1):
        return False
    for i in xrange(2, x):
        if x % i == 0:
            break
    else :
        return True
    return False

# 请在下面完善题目中的代码功能
a = input("输入：")
while a:
    if a == 0:
        print("bye bye")
        exit()
    else:
        if is_prime(a):
            print("yes")
        else:
            print("no")
    a = input("输入：")
```
